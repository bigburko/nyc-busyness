```python


# Load the dataset
#df = pd.read_csv("/Users/jennifercasavantes/Downloads/Alone_or_in_Any_Combination.csv")

# Convert all columns to numeric where possible
df = df.apply(pd.to_numeric, errors='ignore')

# Step 1: Identify columns that contain each category
white_cols = [col for col in df.columns if 'White ' in col]
black_cols = [col for col in df.columns if 'Black or African American' in col]
asian_cols = [col for col in df.columns if 'Asian' in col]
hispanic_cols = [col for col in df.columns if 'Hispanic' in col]

# Step 2: Sum columns by group
df['White '] = df[white_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Black'] = df[black_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Asian'] = df[asian_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)
df['Hispanic'] = df[hispanic_cols].apply(pd.to_numeric, errors='coerce').sum(axis=1)

# Step 3: Use total population column (you can confirm exact name here)
pop_col = [col for col in df.columns if "Pop" in col or "Total" in col][0]
df['Pop'] = pd.to_numeric(df[pop_col], errors='coerce')

# Step 4: Compute percentages
df['% White'] = df['White '] / df['Pop']
df['% Black'] = df['Black'] / df['Pop']
df['% Asian'] = df['Asian'] / df['Pop']
df['% Hispanic'] = df['Hispanic'] / df['Pop']

# Step 5: Keep relevant results
demo_tract = df[['White ', 'Black', 'Asian', 'Hispanic', 'Pop', '% White', '% Black', '% Asian', '% Hispanic']]

# Display first rows
print(demo_tract.head())




```
